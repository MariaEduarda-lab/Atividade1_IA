```python
## O código a seguir trocará a coluna "date_added" que é um objeto para datetime.texto em itálico
```

```python
df['date_added'] = pd.to_datetime(df['date_added'])
```

```python
df.head()
```

```python
## Acima, foi possível entender que a coluna "date_added" teve seus valores mudados, e dessa forma, as notações das datas foram alteradas, para ficar coeso com o padrão de ano-mês-data e com valores cheios, ou seja, com o "preenchimento de zero à esquerda".
```

```python
## Abaixo será apresentada uma criação de novas colunas, nomeadas "month_added" e "year_added", que serão responsáveis por destrinchar as datas apresentadas, nessas duas novas colunas, apresentando, a partir daí, valores de apenas mês e ano, em cada coluna, respectivamente.
```

```python
df['month_added'] = df['date_added'].dt.month
df['year_added'] = df['date_added'].dt.year
```

```python
df.head()
```

```python
## Essa tabela acima apreseta as novas colunas criadas, a partir da coluna "release-year", sendo elas de "month_added" e "year_added", respectivamente, que possibilitará fazer análises relacionadas a esses campos de maneira masi precisa em relação a mês e ano de maior visibilidade, respectivamnete.
```

```python
df['type'].value_counts()
```

```python
## Acima, é possível entender que existem dois tipos de conteúdo assistido, sendo eles "movie" e "tv show", respectivamente, e dentre eles, movie é o mais assistido.
```

```python
## Abaixo será filtrado os filmes com mais de 100 minutos lançados a partir de 2020
```

```python
filmes = df[df['type'] == 'Movie'].copy()

filmes['duration_int'] = filmes['duration'].str.extract('(\d+)').astype(float)

filmes_100min_2020 = filmes[
    (filmes['duration_int'] > 100) &
    (filmes['release_year'] >= 2020)
]

filmes_100min_2020
```

```python
## É perceptível acima, um corte dos filmes com mais de 100 minutos, lançados a partir de 2020, utilizando o corte de ano, coluna adicionada anteriormente "year_added". Há 339 linhas que apresentam essa respotas, mas são exibidas apenas 10.
```

```python
## Abaixo, a coluna "duration" será separada em duas colunas "número" e "unidade", respectivamente.
```

```python
df[['duration_int', 'duration_unit']] = df['duration'].str.extract(r'(\d+)\s*(\w+)')
df['duration_int'] = df['duration_int'].astype(float)
```

```python
df.head()
```

```python
## Analisando o que foi feito acima, e o resultado disso, em algumas linhas, é possível entender que a separação feita para entender que existia dois parâmetros de avaliação, tanto em temporada quanto em minutos. Dessa forma a tabela faz mais sentido.
```

```python
## Abaixo será criada a coluna "seasons_int" com valor numérico, e ela indicará para as linhas de TV show, quantas temporadas existem.
```

```python
df['seasons_int'] = df.apply(
    lambda row: row['duration_int'] if row['type'] == 'TV Show' else None,
    axis=1
)
```

```python
df.head()
```

```python
## Acima, é possível perceber a criação de uma nova coluna, substituindo a análise feita em duas colunas anteriormente, sendo elas "duration_int" que apresenta a quantidade de temporadas no caso de TV show, e duration_unit, que apresenta se a medição é feita em temporadas ou minutos.
```

```python
## Abaixo o código apresentará sa sére com mais temporadas e filme com maior duração
```

```python
# Série com mais temporadas
df[df['type'] == 'TV Show'].sort_values(by='duration_int', ascending=False).head(1)
```

```python
# Filme com maior duração
df[df['type'] == 'Movie'].sort_values(by='duration_int', ascending=False).head(1)
```
